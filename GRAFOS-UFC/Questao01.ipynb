```python
from heapq import heappop, heappush

# Implementação do algoritmo de Prim, que vai me retorna uma lista de arestas que fazem parte da MST
def prim(grafo, origem):
    num_vertices = len(grafo)  #total de vertices
    arestas = [(-1, -1)] * num_vertices  # aresta[v] = (pesoDaAresta(u->v), u)
    visitado = [False] * num_vertices # visitado[v] = True se v já foi visitado
    arestas[origem] = (0, -1) # A raiz da MST não tem ninguém antes dela, então o pai é -1
    fila_prioridade = [(0, origem)] # Fila de prioridade que vai conter os pares (peso, vértice)

    print(visitado)

    while fila_prioridade: # Enquanto a fila não estiver vazia
        _, v = heappop(fila_prioridade) # Desempilha o vértice v da fila de prioridade

        if visitado[v]: # Se o vértice já foi visitado, não faça nada
            continue
        
        visitado[v] = True # Marca o vértice como visitado


        for u, peso in grafo[v]: # Percorre todos os vizinhos u de v
            if not visitado[u] and (arestas[u][0] == -1 or peso < arestas[u][0]): # Se u não foi visitado e o peso da aresta (u, v) é menor que o peso da aresta (u, arestas[u])
                arestas[u] = (peso, v) # Atualiza a aresta (u, arestas[u])
                heappush(fila_prioridade, (peso, u)) # Adiciona o par (peso, u) na fila de prioridade

    return arestas # Retorna a lista de arestas que fazem parte da MST


# Definição do grafo não direcionado representado por uma lista de adjacências
grafo = [[] for _ in range(9)] 

# Adicionando as arestas ao grafo
def add_aresta(v, u, w):
    grafo[v].append((u, w))
    grafo[u].append((v, w))  # Considera que o grafo é não direcionado

add_aresta(0, 1, 4)
add_aresta(0, 7, 8)

add_aresta(1, 2, 8)
add_aresta(1, 7, 11)

add_aresta(2, 3, 7)
add_aresta(2, 5, 4)
add_aresta(2, 8, 2)

add_aresta(3, 4, 9)
add_aresta(3, 5, 14)

add_aresta(4, 5, 10)

add_aresta(5, 6, 2)

add_aresta(6, 8, 6)
add_aresta(6, 7, 1)

add_aresta(7, 8, 7)

#print do grafo
""" print("graph", grafo) """


# Chamando a função prim para encontrar a árvore geradora mínima 
mst = prim(grafo, 0)

# Imprimindo a árvore geradora mínima
print("Árvore Geradora Mínima:")
for v in range(1, len(mst)): # Para cada vértice v da MST
    peso, u = mst[v] # Peso da aresta (u, v)
    if v == 8:
        print(f"Aresta {u} <-> {v} com peso {peso}")
        break
    elif u != -1:
        print(f"Aresta {u} <-> {v} com peso {peso}")
        
```

```python
def encontrar_caminho_minimo(vInicial, vDestino, arvore_geradora):
    if(vInicial > vDestino): # Se o vértice inicial for maior que o vértice destino, troca os vértices, isso pq to fazendo o caminho de volta
        vInicial, vDestino = vDestino, vInicial

    caminho = []
    peso_total = 0

    v = vDestino
    while v != vInicial: # Enquanto o vértice atual não for o vértice inicial
        weight, u = arvore_geradora[v] # Peso da aresta (u, v)
        caminho.append((u, v, weight)) # Adiciona a aresta (u, v) no caminho mínimo
        peso_total += weight # Soma o peso da aresta (u, v) no peso total
        v = u # Vai para o prox, so vai parar quando o vertice atual for o vertice inicial

    print(caminho)
    caminho.reverse() # Inverte o caminho mínimo
    return caminho, peso_total

vInicial = 8  # Vértice inicial modificado
vDestino = 0  # Vértice destino modificado

caminho_minimo, peso_total = encontrar_caminho_minimo(vInicial, vDestino, mst)

print("Caminho mínimo entre os vértices:")
for u, v, weight in caminho_minimo:
    print(f"Aresta {u} <-> {v} com peso {weight}")

print("Peso total:", peso_total)
```
